```python
data = data.drop(columns=['Overall','ID'])
corr_matrix = data.corr()
corr_matrix
```

**Luego de hacer un analisis de correlación entre las variables, se seleccionan las 30 variables con un mayor valor de correlación para poderlas utilizar en nuestro modelo**

```python

# Seleccionar las 5 variables con las correlaciones más altas en orden descendente
top_corr_vars = pd.DataFrame(corr_matrix['Potential'].nlargest(28))
top_corr_vars

```

**De nuestro dataset original, seleccionamos unicamente las 30 variables que identificamos de mayor importancia en el paso anterior y descartamos el resto.**

```python
columns = top_corr_vars.index.tolist()
data = data.loc[:, columns]
data
```

**Finalmente para asegurarno, podemos verificar si hay algun dato nulo en nuestro nuevo dataset.**

```python
conteo_de_nulos = data.isnull().sum()
null_len = len(conteo_de_nulos[conteo_de_nulos > 0])

total_rows = len(data.index)
null_rows = len(data.index[data.isnull().any(axis=1)])
null_rows_percentage = (null_rows / total_rows) * 100

print('Se encontraron {} datos nulos en el dataset.'.format(null_len))
print('Lo que indica una distribución del {}%  de datos nulos en el dataset.'.format(null_rows_percentage))
```

```python
#Obtenemos las columnas que tienen al menos 1 valor nulo y descartamos el resto.

columnas_con_nulos = conteo_de_nulos[conteo_de_nulos>0].index.tolist()
```

**#Obtenemos la distribución de clases para saber todos los valores disintos en la columna Potential y su conteo**

Este paso lo hacemos para poder reemplazar los valores nulos de cada columna con el promedio de la misma columna, pero tomando en cuenta las filas que tienen un mismo valor en la columna 'Potential'

```python
classes_count = pd.value_counts(data['Potential'])
classes_count
```

**Filtramos y conservamos unicamente los indices del dataframe anterior, esto para obtener las diferentes clases de la columna 'Potential**'

```python
class_distribution = classes_count.index.tolist()
print("Distribución de clases:\n", class_distribution)
```

```python
data
```

```python
for Class in classes_count.index.tolist():
    for column in columnas_con_nulos:
        data_temp = data[data['Potential'] == Class]
        cb_mean = round(data_temp[column].mean(),0)
        data.loc[data['Potential'] == Class, column] = data.loc[data['Potential'] == Class, column].fillna(cb_mean)
```

```python
new_order = list(data.columns[1:]) + [data.columns[0]]
data = data.reindex(columns=new_order)
```

```python
data
```

```python
columnas_con_nulos = data.isnull().sum()
null_len = len(columnas_con_nulos[columnas_con_nulos > 0])
print('Se encontraron {} datos nulos en el dataset.'.format(null_len))
```

<h2>Normalización<h2>

```python
X = data.iloc[:, 1:].values
y = data.iloc[:, 0].values.reshape(-1, 1) # Blue wins 
```

```python
from sklearn.preprocessing import StandardScaler
normalizador = StandardScaler()
X = normalizador.fit_transform(X)
X
```

## Split de datos

```python
X_entreno, X_prueba, y_entreno, y_prueba = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

Luego de hacer varias pruebas, se determino que los hiperparámetros, max_depth y min_split

```python
from DesitionTreeClasifier import DesicionTree
tree = DesicionTree(max_depth=8, min_split=4)
tree.fit(X_entreno, y_entreno)
```

```python
y_pred = tree.predict(X_prueba)
from sklearn.metrics import precision_score, accuracy_score
print('Acuracy', accuracy_score(y_prueba, y_pred))
```
